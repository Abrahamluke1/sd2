```python
from tabulate import tabulate
from datetime import date, datetime
from dateutil.relativedelta import *


# function to handle eos error
def handele_eos(input_value):
    try:
        value = input(input_value)
        return value
    except EOFError as er:
        return None


# function to calculate the rating
def calculate_rating(result):
    overall_rating = float(result * 100 / 30)
    return overall_rating


# function to calculate the salary
def calculate_salary(overall_rate):
    if overall_rate >= 80:
        return "1000"
    elif overall_rate >= 60 and overall_rate < 80:
        return "1000 700"
    elif overall_rate >= 45 and overall_rate < 60:
        return "700 500"
    elif overall_rate >= 30 and overall_rate < 45:
        return "500 400"
    else:
        return "400"


# function to validate date format
def validate_dob(date_string):
    flag = True
    try:
        dateObect = datetime.strptime(date_string, '%Y-%m-%d').date()
    except Exception as e:
        flag = False
    return flag

def manage_player_data(data):
    # sort player details lambda function
    data.sort(key=lambda x: x[0])
    # print table using tabulate package
    print(tabulate(data, headers=["UID", "Name", "D.o.B", "Age", "Score", "Salary Range"]))
    # writing file to players.txt file
    with open('players.txt', 'w') as f:
        f.write(tabulate(data, headers=["UID", "Name", "D.o.B", "Age", "Score", "Salary Range"]))


def main():
    player_data_table = []
    # running a loop to get data of players
    while True:
        id = handele_eos("Enter player id ")
        if id == 'end':
            break
        player_name = handele_eos("Enter player Name --> ")
        player_dob = handele_eos("Enter DOB --> ")
        if not validate_dob(player_dob):
            print("The date you entered was invalid., should be YYYY-MM-DD")
            break

        # age calculating using dob of the player
        days_in_year = 365.2425 
        birthdate = datetime.strptime(player_dob, '%Y-%m-%d').date()
        age = int((date.today() - birthdate).days / days_in_year)
        print(age)

        total_score = 0
        one_player_data = [id, player_name, player_dob, age]
        skills = ['Speed', 'Shooting', 'Passing', 'Defending', 'Dribbling', 'Physicality']

        # looping through the skills to get values for each skill
        for experiance_rating in skills:
            try:
                skill_data = int(input("Enter the " + experiance_rating + "--> "))
                if skill_data in range(0,6):
                    total_score += skill_data
                    continue
                print("The rating you entered was invalid.")
                break
            except:
                print("The rating you entered was invalid.")

            
        total_rating = calculate_rating(total_score)
        salary_range = calculate_salary(total_rating)

        #adding data to dict to print the table
        one_player_data.append(total_rating)
        one_player_data.append(salary_range)
        player_data_table.append(one_player_data)

    # function to sort, print and save player data
    manage_player_data(player_data_table)

if (__name__)== '__main__':
    main()
```
